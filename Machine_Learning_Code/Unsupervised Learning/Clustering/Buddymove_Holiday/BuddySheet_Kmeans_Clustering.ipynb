# K-Means Clustering

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
df=pd.read_csv('buddymove_holidayiq.csv')
df.head()
```

```
X=df.iloc[:,1:].values
```

# Feature Scalling

```
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X=sc.fit_transform(X)
```

## Using the elbow method to find the optimal number of clusters

```
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
  km=KMeans(n_clusters=i,init='k-means++',n_init=10,random_state=10)
  km.fit(X)
  wcss.append(km.inertia_)

plt.plot(range(1,11),wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Cluster')
plt.ylabel('Distance')
plt.legend()
plt.show()
```

## Training the K-Means model on the dataset

```
kmeans=KMeans(n_clusters=2,init='k-means++',n_init=10,random_state=43)
y_kmeans=kmeans.fit_predict(X)
centroids=kmeans.cluster_centers_
y_kmeans
```

```
df['cluster']=y_kmeans
df.head()
```

**Calculate No of observation in Cluster bold text**

```
Cluster_1=len(df[df['cluster']==0])
Cluster_2=len(df[df['cluster']==1])
print('No of Cluster 1 is ',Cluster_1)
print('No of Cluster 2 is ',Cluster_2)
```
